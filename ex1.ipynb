```python
def p(x):
    """définir la fonction polynome  p(x)"""
    return (x-1)**6
import numpy as np


for x in np.arange (0.998,1.002,0.001):
    
        print('p({})={} '.format(x,p(x)))

def f(x):
    """définir la fonction polynome  f(x)"""
    return x**6 - 6*x**5 + 15*x**4 - 20*x**3 + 15*x**2 - 6*x + 1

import numpy as np
for x in np.arange (0.998,1.002,0.001):
    
        print('p({})={} '.format(x,f(x)))

import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0.998,1.002, 70)
plt.plot(x,f(x),'r',label="f(x)")
plt.plot(x,p(x),'b*',label="p(x)")
plt.xlabel('X')
plt.ylabel('f(x)')
plt.legend()
plt.title('Solution dans [0.998,1.002]')
plt.grid()
```
